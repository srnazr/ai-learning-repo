```python
import numpy as np 
import pandas as pd 

print("All modules imported successfully!")

pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("./telecom_churn_data.csv")
print("Data imported successfully!")

df.head()
```

```python
print(df.shape)
```

##### OUTPUTS GENERAL INFO ABOUT THE DATA FRAME

```python
print(df.info())
```

```python
print(df.columns.tolist())
```

##### CASTING A COLUMN'S DATA TYPE

```python
df["Churn"] = df["Churn"].astype("int64")

print(df.info())
```

```python
df.describe()
```

##### TO SEE STATISTICS OF NON-NUMERIC VALUES

```python
df.describe(include="object")

# A list of non-numeric columns can also be passed: include=["object", "bool"]
```

##### TO GET THE COUNTS OF NON-NUMERIC VALUES (EX: BOOL)

```python
print(df["Churn"].value_counts())

# To return the counts as fractions we can also use:
print(f"\n{df['Churn'].value_counts(normalize=True)}")
```

##### SORTING VALUES

```python
# Sort using 1 column
df.sort_values(by="Total day charge", ascending=False).head()
```

```python
#Sort using multiple columns
df.sort_values(by=["Churn", "Total day charge"], ascending=[True, False]).head()
```

##### GETTING THE MEAN

```python
print(df["Churn"].mean())
```
